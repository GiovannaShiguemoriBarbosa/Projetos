```python
coastal_tourism = df[df['sector'] == 'Coastal tourism']
filtered_df = coastal_tourism[
    (coastal_tourism['country_name'] == 'Italy') &
    (coastal_tourism['indicator_name'] == 'Personnel costs - million euro')
]

grouped_df = filtered_df.groupby(pd.to_datetime(filtered_df['ref_date']).dt.year)['value'].sum().reset_index()
grouped_df.plot(x='ref_date', y='value', kind='bar', title='Personnel costs in the Italian coastal tourism sector')

```
